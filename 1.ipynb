## Работа со строками

Для начала стоить сказать, что строка в пайтоне это список символов,отсюда следует, что мы можем обратиться к символу по индексу

```python
name = "Артём"
print(name[0])  # Выводит первую букву имени
```

### Срезы
Срезы строки, это взятие части строки по индексам, срезы мы берем через[a:b], где a-индекс от которого мы берем срез, b-индекс+1 до которого идёт срез, еслит мы отсавляем a-пустым, то у нас будет срез начинаться с 1 элемента, если b-пустым, то он будет идти до полседнего элемента

```python
name = "Артём"
print(name[1:3])  # Выводит вторую и третью буквы имени
my_name = name[:4]  # Сохраняет первые четыре буквы имени в переменную my_name
print(my_name)
print(name[2:])  # Выводит все буквы имени, начиная с третьей
```

### Методы upper(), lower() и функция len()
Метод upper() переводит символы нашей строки в верхний регистр и получившеюся строку возвращает в переменную или передаёт в функцию, lower() тоже самое, только нижний регистр. Функция len() считает длину нашей строки(строка передаваемое значение), и возвращет длину строки в передменную или передаёт в другую функцию.

```python
name = "Артём"
u_name = name.upper()  # Преобразует имя в верхний регистр
l_name = name.lower()  # Преобразует имя в нижний регистр
print(u_name)
print(l_name)
print(name.upper())  # Выводит имя в верхнем регистре без сохранения в переменную
print(name.lower())  # Выводит имя в нижнем регистре без сохранения в переменную
```

# Условный оператор if.

Условный оператор if проверяет истинность выражения, те. принимает булевое значение(bool) и в случае если оно содержит True(Истина), то будет выполнятся блок кода для if, те. находящийся с разницей в 1 табуляцию.

```python
a = 7
b = 14
if a < b:
    print(f"a = {a} меньше b = {b}, но мы увеличиили a и теперь он равен {a+(b-a)+1}")
    a += (b-a) + 1
```

### Условный оператор else

Блок кода для else выполниться в том случае, если условие в if False(Ложь)

```python
a = 14
b = 7
if a < b:
    print(f"a = {a} меньше b = {b}, но мы увеличиили a и теперь он равен {a+(b-a)+1}")
    a += (b-a) + 1
else:
    print(f"a = {a} больше b = {b}, но мы увеличиили b и теперь он равен {b+(a-b)+1}")
    b += (a-b) + 1
```

### Условный оператор elif

elif выполняется только в том случае, если у нас if неверен, а он в отличие от else умеет проверять условия.
у нас может быть много elif, они всегда идут после if. Так же после elif мы можем ставить else, которое будет выполняться только тогда, когда у нас и if и elif неверно.

```python
# Запрашиваем два числа у пользователя
num1 = int(input("Введите первое число: "))
num2 = int(input("Введите второе число: "))

# Сравниваем числа
if num1 > num2:
    print("Первое число больше.")
elif num1 < num2:
    print("Второе число больше.")
else:
    print("Числа равны.")
```

### Логические операторы.

Логические операторы необходимы для обьеденения нескольких операторов в 1 логическое выражение<br>
and - оператор который проверяет условие слева и справо от него и если хотя-бы 1 ложь, то он вернет ложь<br>
or - оператор который проверяет условие слева и справо от него и если хотя-бы 1 истина, то он вернет истина<br>
not - оператор который делает значение оператора который стоит перед ним обратным, те. если была ложь, станет истина, если была истина, станет ложь

```python
age = 21
if age > 18 and age < 30:
    print("Вы молодой человек")
   
day = "Понедельник"
if day == "Понедельник" or day == "Вторник":
    print("Сегодня рабочий день")

is_ready = False
if not is_ready:
    print("Мы не готовы")
```

## Сложные условные операторы

### Вложенные условия
Это процесс, когда логические конструкции находятся внутри других условий

```python
# Данные игрока
has_key = True  # Есть ли у игрока золотой ключ?
level = 5       # Уровень игрока

# 1-й шаг: Проверяем самое главное — есть ли ключ
if has_key == True:
    print("Ключ подошел! Дверь открывается...")
    
    # 2-й шаг: Проверяем, достаточно ли игрок силен
    if level >= 10:
        print("Поздравляем! Вы вошли в секретную комнату босса.")
    else:
        print("Внутри слишком опасно! Ваш уровень ниже 10, вас не пустили.")

else:
    print("Дверь заперта. Сначала найдите золотой ключ!")
```
