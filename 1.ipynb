# Python
## Переменные и типы данных

Мы можем создать ячейку в памяти для хранения некоторых данных - переменная, для создания переменной нам требуется написать ее название, знак присваивания(=) и значение которым мы её инициализируем
Пример:

```python
name = "Артём"
age = 21
```

### Типы данных
У нас есть разные типы данных, вот основные<br>
Целочисленные(int):

```python
number = 10
number2 = 20    
```

Строки(str) по сути строка это список символов(char):

```python
name = "Артём"
sec_name = "Стрелов"
```

Числа с плавающей запятой(float):

```python
p = 3.14
height = 10.5
```

Булевы(bool):

```python
is_ready = True
is_open = False
```

## Ввод и вывод данных

### Ввод данных
Перед тем как говорить про ввод данных, стоит сказать, что мы занимаемся функциональным программированием, тоесть мы используем функции для написания программ, и все функции работаю следующим образом:<br>
1.У функии есть название<br>
2.У функции есть передаваемые значения(необязательно, но место для передаваемых в функцию значения оставлять обязательно), место для передаваемых значений это скобки после функции<br>
3.У функции есть возращаемые значения(необязательно, но если есть, то нам надо куда либо присвоить значение, в переменную или передать другой функции)<br>
Так вот для вывода данных используется функция print()<-внутри скобочек пишем передаваемое значение, которо надо вывести<br>
Пример:

```python
age = 21
print("Hello, world!")
print(age)
```
